```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/kaggle/input/datasets/organizations/uciml/iris/Iris.csv')
"""
Scatter Plot
A scatter plot is used to show the relationship between two numerical variables.
The scatter plot shows how sepal length and sepal width are related across different species of flowers.
Some species are clearly separated
Others overlap slightly

"""
plt.figure()
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
"""
Histogram
A histogram is used to show the distribution of a single variable.
The histogram displays the distribution of petal length values in the dataset. 
It shows how frequently different ranges of petal length occur.
Data is not perfectly uniform
Some ranges have more values (higher bars)
Helps understand spread and concentration
"""

plt.figure()
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Sepal Length by Species')
plt.show()
```

```python
"""
Box Plot
A box plot is used to show the spread of data and detect outliers.
The box plot summarizes the distribution of sepal length for each species. 
It shows the median, quartiles, and possible outliers. 
This visualization helps compare how sepal length varies across different species.
Median differs between species
Some species have wider spread
Outliers (if any) are visible as points
"""
plt.figure()
sns.histplot(df['PetalLengthCm'], bins=20, kde=True)
plt.title('Distribution of Petal Length')
plt.show()
```
